# Fine-Tuning Experiments

Этот ноутбук нужен для демонстрации этапа экспериментов:
- проверка структуры датасета;
- запуск обучения двух моделей;
- подбор гиперпараметров;
- экспорт лучшей модели в ONNX;
- просмотр результатов.


```python
from pathlib import Path
import json
import pandas as pd

PROJECT_ROOT = Path.cwd().resolve().parents[0] if Path.cwd().name == "experiments" else Path.cwd()
PROJECT_ROOT
```

```python
DATA_ROOT = PROJECT_ROOT / "data" / "raw" / "my_dataset"
for split in ["train", "val", "test"]:
    split_path = DATA_ROOT / split
    print(f"\n{split_path}:")
    for cls_dir in sorted(split_path.iterdir()):
        if cls_dir.is_dir():
            print(cls_dir.name, len(list(cls_dir.glob("*"))))
```

## Обучение двух моделей

```python
import subprocess
import sys

cmd = [
    sys.executable,
    str(PROJECT_ROOT / "experiments" / "train.py"),
    "--data-dir", "data/raw/my_dataset",
    "--epochs", "8",
    "--freeze-epochs", "3",
]
subprocess.run(cmd, cwd=PROJECT_ROOT, check=True)
```

## Просмотр таблицы результатов

```python
results_path = PROJECT_ROOT / "models" / "results.csv"
pd.read_csv(results_path)
```

## Подбор гиперпараметров

```python
cmd = [
    sys.executable,
    str(PROJECT_ROOT / "experiments" / "train.py"),
    "--data-dir", "data/raw/my_dataset",
    "--run-sweep",
    "--epochs", "6",
    "--freeze-epochs", "2",
]
subprocess.run(cmd, cwd=PROJECT_ROOT, check=True)
```

```python
sweep_path = PROJECT_ROOT / "models" / "sweep_results.csv"
pd.read_csv(sweep_path)
```

## Экспорт лучшей модели в ONNX

```python
cmd = [
    sys.executable,
    str(PROJECT_ROOT / "experiments" / "train.py"),
    "--data-dir", "data/raw/my_dataset",
    "--epochs", "8",
    "--freeze-epochs", "3",
    "--export-onnx",
    "--onnx-file-name", "best_model.onnx",
]
subprocess.run(cmd, cwd=PROJECT_ROOT, check=True)
```

```python
summary_path = PROJECT_ROOT / "models" / "best_run_summary.json"
json.loads(summary_path.read_text(encoding="utf-8"))
```
